## ADANI PORTS DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./datasets/Experiment1/ADANIPORTS.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
last_six_mon_df = data[data['Date'] >= data['Date'].max() - pd.DateOffset(months=6)]
```

```python
last_six_mon_df['Trades'] = last_six_mon_df['Trades'].bfill()
```

```python
last_six_mon_df.isnull().sum()
```

```python
last_six_mon_df.describe()
```

```python
plt.figure(figsize=(12, 6))  
plt.plot(last_six_mon_df['Date'], last_six_mon_df['Open'], label='Open Price', color='blue', marker='o')  
plt.plot(last_six_mon_df['Date'], last_six_mon_df['Close'], label='Close Price', color='green', marker='o') 

plt.title('Opening and Closing Prices Over Time', fontsize=16)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Price', fontsize=14)

plt.legend()

plt.xticks(rotation=45)

plt.tight_layout() 
plt.show()
```

### How much has data differed

```python
last_six_mon_df['Daily Volatility'] = last_six_mon_df['High'] - last_six_mon_df['Low']
plt.figure(figsize=(12, 6))
plt.plot(last_six_mon_df['Date'], last_six_mon_df['Daily Volatility'], label='Daily Volatility', color='red')
plt.title('Daily Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

### How much the investors are interesed in buying this share over time

```python
plt.figure(figsize=(12, 6))
plt.plot(last_six_mon_df['Date'], last_six_mon_df['%Deliverble'], label='% Deliverable', color='green')
plt.title('% Deliverable Over Time')
plt.xlabel('Date')
plt.ylabel('% Deliverable')
plt.legend()
plt.show()
```
